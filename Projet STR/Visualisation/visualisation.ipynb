Mise en place des modules utiles

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Ouverture du fichier csv

```python
df = pd.read_csv("../log.csv", names=["time_per_frame", "reaction_time", "detection_time"])
```

Affichage des Ã©volution des valeurs

```python
fps_df = df.loc[df["time_per_frame"] > 0, ["time_per_frame"]]
fps_df["fps_goal"] = 1/60
plt.plot(fps_df["time_per_frame"])
plt.plot(fps_df["fps_goal"])
```

```python
rec_df = df.loc[df["reaction_time"] > 0, ["reaction_time"]]
rec_df["rec_goal"] = 0.273
plt.plot(rec_df["reaction_time"])
plt.plot(rec_df["rec_goal"])
```

```python
det_df = df.loc[df["detection_time"] > 0, ["detection_time"]]
det_df["det_goal"] = 1
plt.plot(det_df["detection_time"])
plt.plot(det_df["det_goal"])
```
